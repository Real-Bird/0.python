### 네이버 뉴스 속보 데이터화
1. 크롤링한 데이터 변수 할당
    - total 리스트 생성
        + for loop에서 total 리스트에 추가
    - total에서 DataFrame 만들기
        + 컬럼명 -> img, link, title, content
2. naver_news.csv

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {"User-Agent":"Mozilla/5.0"}
url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"
html = requests.get(url, headers = headers).text
soup = BeautifulSoup(html, "html.parser")
type06 = soup.find("ul", {"class" : "type06_headline"})
dl = type06.find_all("dl")
total = [[] for i in range(len(dl))]
for no, item2 in enumerate(dl):
    try:
        img = item2.find("dt",{"class" : "photo"}).find("img")
        print("img = ", img["src"])
        link = item2.find("dt",{"class": ""}).find("a")
        print("link = ", link["href"])
        title = link.text.replace("\t","")[1:len(link.text)+1]
        print("title = ", title)
        content = item2.find("dd").text.replace("\t","").replace("\n","")\
        .split("...")[0]
        print("content = ", content)
        print("-"*30)
        total[no].append(img["src"])
        total[no].append(link["href"])
        total[no].append(title.replace("\t","").replace("\n","").replace("\r","")
                         .split("...")[0])
        total[no].append(content)
    except:
        print("No data = ", no)

pd.set_option('display.max.colwidth', 30)
news = pd.DataFrame(total, columns=["img", "link", "title", "content"])
news.to_csv("naver_news.csv",encoding="cp949",index=False)
pd.read_csv("naver_news.csv", encoding="cp949")
```

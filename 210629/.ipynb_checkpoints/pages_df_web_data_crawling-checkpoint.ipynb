### 네이버 뉴스 속보 함수로 여러 페이지 작업
1. 날짜는 20210628 ~ 20210629 2일...
    - 날짜별 page는 2개만
2. 4개의 page 수집
3. DataFrame 만들기
    - img, link, title, content, date, page
    - total DataFrame 누적
4. total DataFrame 컬럼 순서 바꾸기
    - ["date", "page", "img", "link", "title", "content"]
    - total.shape

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
headers = {"User-Agent":"Mozilla/5.0"}
```

```python
def get_page(url):
    html = requests.get(url, headers = headers)
    soup = BeautifulSoup(html, "html.parser")
    type06 = soup.find("ul", {"class" : "type06_headline"})
    dl = type06.find_all("dl")
    return dl
```

```python
def page_to_items(dl):
    total = []
    for no, item2 in enumerate(dl):
        try:
            img = item2.find("dt",{"class" : "photo"}).find("img")
            link = item2.find("dt",{"class": ""}).find("a")
            content = item2.find("dd")
            
            img_src = img["src"]
            link_href = link["href"]
            title = link.text.replace("\t","").replace("\r\n","").strip()
            content_text = content.text.replace("\n","").split("...")[0]
            
            total.append([img_src,link_href, title, content_text])

        except:
            print("No data = ", no)
    
    return total
```

```python
# https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001&date=20210629&page=1 
news_date = ["20210628","20210629"]
news_page = ["1", "2"]

base_url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"

init = 0
for date in news_date:
    for page in news_page:
        url = base_url + "&date=" + date + "&page=" + page
        get_dl = get_page(url)
        total_news = page_to_items(get_dl)
        news_df = pd.DataFrame(total_news, columns=["img", "link", "title", "content"])
        news_df["date"] = date
        news_df["page"] = page
        
        if init == 0:
            total_df = news_df.copy()
            init = 1
        else:
            total_df = pd.concat([total_df, news_df], ignore_index = True)
            

total_df = total_df[["date", "page", "img", "link", "title", "content"]]
total_df.to_csv("naver_news3.csv",index=False)
pd.read_csv("naver_news3.csv")
```

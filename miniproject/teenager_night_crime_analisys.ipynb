```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import csv
f = open("crime_teenager.csv")
data = csv.reader(f)
next(data)
col = ['지역', '살인', '강도', '강간', '절도', '폭행']
```

```python
teen_crime = pd.DataFrame(data, columns=col)
```

```python
teen_crime
```

```python
teen_crime.describe()
```

```python
# dtype => int64로 변경 후 "지역" 기준 내림차순 정렬
teen_crime = teen_crime.sort_values(by="지역")
teen_crime
```

```python
location = teen_crime["지역"]
murder = teen_crime["살인"]
robbery = teen_crime["강도"]
rape = teen_crime["강간"]
thief = teen_crime["절도"]
violence = teen_crime["폭행"]
crimes = [murder, robbery, rape, thief, violence]
```

```python
plt.rc("font", family="Malgun Gothic")
for i in crimes:
    plt.bar(location, i)
    plt.legend()
    plt.show()
```

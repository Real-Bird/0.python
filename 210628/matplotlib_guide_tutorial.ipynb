```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1,1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, {"marker": "x"})
my_plotter(ax2, data1, data2, {"marker": "o"})
```

```python
fig, ax = plt.subplots(2,2)
my_plotter(ax[0,0], data1, data2, {"marker": "x"})
my_plotter(ax[1,1], data1, data2, {"marker": "o"})
```

```python

days = tips.day.value_counts()
index = days.index
values = days.values

fig1, ax1 = plt.subplots()
ax1.pie(values, explode = (0, 0.1, 0, 0), labels=index, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
def pie_graph(feature):
    index = tips[feature].value_counts().index
    values = tips[feature].value_counts().values
    fig1, ax1 = plt.subplots()
    ax1.pie(values, labels=index, autopct='%1.1f%%',
        shadow=True, startangle=90)
    ax1.axis('equal')

    plt.show()
    return
```

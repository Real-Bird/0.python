```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("auto-mpg.csv", header=None)
df.columns = ["mpg","cylinders","displacement","horsepower","weight",
             "acceleration","model year","origin","name"]
```

```python
df.head()
```

```python
pd.set_option("display.max_columns", 10)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["horsepower"].unique()
```

```python
df["horsepower"].replace("?", np.nan, inplace=True)
```

```python
df.dropna(subset=["horsepower"],axis=0,inplace=True)
```

```python
df["horsepower"] = df["horsepower"].astype("float")
df.describe()
```

```python
ndf = df[["mpg","cylinders","horsepower","weight"]]
ndf.head()
```

```python
ndf.plot(kind="scatter",x="weight",y="mpg",c="gold",s=10,figsize=(10,5))
plt.show()
plt.close()
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
sns.regplot(x="weight", y="mpg",data=ndf, ax=ax1, color="red")
sns.regplot(x="weight", y="mpg",data=ndf, ax=ax2, color="orange",fit_reg=False)
plt.show()
plt.close()
```

```python
sns.jointplot(x="weight",y="mpg",data=ndf)
sns.jointplot(x="weight",y="mpg",kind="reg",data=ndf)
plt.show()
plt.close()
```

```python
grid_ndf=sns.pairplot(ndf)
plt.show()
plt.close()
```

```python
x = ndf[["weight"]]
y = ndf[["mpg"]]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,\
                    y, test_size=0.3, random_state=10)

print("train data 개수 :", len(x_train))
print("test data 개수 : ", len(x_test))
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train, y_train)

r_square = lr.score(x_test, y_test)
print(r_square)
```

```python
print("기울기 a : ", lr.coef_)
print("y절편 b : ", lr.intercept_)
```

```python
y_hat = lr.predict(x)

plt.figure(figsize=(10,5))
ax1 = sns.distplot(y, hist=False, label="y")
ax2 = sns.distplot(y_hat, hist=False, label="y_hat", ax=ax1)
plt.show()
plt.close()
```

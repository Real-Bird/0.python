```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
import mglearn
import matplotlib.pyplot as plt
```

```python
nc = pd.read_csv("../miniproject/nightlife_crime.csv", encoding="cp949")
```

```python
nc_df = pd.DataFrame(nc)
nc_df
```

```python
# 청소년 5대 범죄 df
df = nc_df.iloc[:,4:9]
```

```python
# clustering
model = KMeans(n_clusters=3)
model = model.fit(df)
print(model.labels_)
```

```python
# 표준화
scaler = StandardScaler()
#scaler = MinMaxScaler()
scaler.fit(df)
X_Scaled = scaler.transform(df)
```

```python
# 2차원 축소
pca = PCA(n_components=2)
pca.fit(X_Scaled)

X_pca = pca.transform(X_Scaled)
                               
print(X_Scaled.shape)
print(X_pca.shape)
```

```python
# 시각화
mglearn.discrete_scatter(X_pca[:,0], X_pca[:,1], model.labels_)
plt.legend(["A","B","C"], loc="best")
plt.xlabel("1 주성분")
plt.ylabel("2 주성분")
plt.show()
```

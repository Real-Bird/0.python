### 유튜브 데이터 수집
1. 화면에서 검색어를 입력
2. selenium으로 키워드 검색
    - source => soup
3. 검색 결과를 수집
    - "h3", {"class":"title-and-badge style-scope ytd-video-renderer"}
    - video type
        + 채널명, 영상제목, 영상url, 조회수, 영상등록날짜
        - video_df
    - "div", {"class":"style-scope ytd-playlist-renderer"}
    - playlist type
        + 채널명, 재생목록제목, 재생목록url
        - playlist_df

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome("c:/python/chromedriver")
```

```python
keyword = input()
url  = "https://www.youtube.com/results?search_query=" + keyword
driver.get(url)
```

```python
source = driver.page_source
```

```python
soup = BeautifulSoup(source, "html.parser")
```

```python
video_type = soup.find_all("div", {"id":"dismissible"})
```

```python
playlist_type = soup.find_all("div", {"class":"style-scope ytd-playlist-renderer"})
```

```python
total_vd = []

for item in video_type:
        ch_name = item.find("tp-yt-paper-tooltip", {"class": "style-scope ytd-channel-name"}).text.replace("\n","").strip()
        print("ch_name = ", ch_name)
        title = item.find("yt-formatted-string",{"class":"style-scope ytd-video-renderer"}).text
        print("title = ", title)
        link = "youtube.com"+item.find("a",{"class":"yt-simple-endpoint style-scope ytd-video-renderer"})["href"]
        print("link = ", link)
        hit = item.find_all("span",{"class" : "style-scope ytd-video-meta-block"})[0].text.split("조회수")[1]
        print("hit = ", hit)
        regdate = item.find_all("span",{"class" : "style-scope ytd-video-meta-block"})[1].text
        print("regdate = ", regdate)
        print("-"*30)
        total_vd.append([ch_name, title, link, hit, regdate])
```

```python
total_pl = []

for item in playlist_type:
        ch_name = item.find("tp-yt-paper-tooltip", {"class": "style-scope ytd-channel-name"}).text.replace("\n","").strip()
        print("ch_name = ", ch_name)
        title = item.find("h3", {"class": "style-scope ytd-playlist-renderer"}).text.replace("\n","").strip()
        print("title = ", title)
        link = "youtube.com"+item.find("a",{"class":"yt-simple-endpoint style-scope ytd-playlist-renderer"})["href"]
        print("link = ",link)
        print("-"*30)
        total_pl.append([ch_name, title, link])
```

```python
import pandas as pd
```

```python
vd_df = pd.DataFrame(total_vd,columns=["ch_name", "title", "link", "hit", "regdate"])
vd_df
```

```python
pl_df = pd.DataFrame(total_pl,columns=["ch_name", "title", "link"])
pl_df
```

```python
vd_df.to_csv("video_type_youtube.csv", index=False)
pl_df.to_csv("playlist_type_youtube.csv", index=False)
```

```python
import option as opt 
import data
import network 
import matplotlib.pyplot as plt
import os 
from tensorflow.keras.models import load_model
from tensorflow.keras import preprocessing
import tensorflow
op = opt.Options()
```

```python
def train_o () :   # 원-핫 이진 벡터 로 학습 
    tokenizer = data.get_word_index()    # 단어 인덱스를 구축
    X_total, Y_total = data.load_one_hot_data(tokenizer)   # 원-핫 이진 벡터 표현
    
    if os.path.exists(op.model_o):
        model = load_model(op.model_o)
    else :
        model = network.make_ann_model()
    
    callback_list = [
        tensorflow.keras.callbacks.EarlyStopping( # 성능 향상이 멈추면 훈련을 중지
        monitor='val_loss', # 모델 검증 정확도를 모니터링
        patience = 5
        ),
        tensorflow.keras.callbacks.ModelCheckpoint( # 에포크마다 현재 가중치를 저장
        filepath=op.model_o, #모델 파일 경로 # “model/model_o_h5“ 와 같이 파일명 직접써도 됨
        monitor="val_loss",
        save_best_only=True
        )
    ]
    
 
    history = model.fit(X_total,
                        Y_total ,
                        epochs=op.num_epochs, 
                        shuffle=True,
                        batch_size=100,
                        validation_split=0.2 )
    
    #model.svae(op.model_o)
    return history
```

```python
def train_e():  # Embedding 으로 학습
    tokenizer = data.get_word_index()  # 단어 인덱스를 구축

    X_total, Y_total = data.load_one_word_seq(tokenizer )

    if os.path.exists(op.model_e):
        model = load_model(op.model_e)
    else :
        model = network.make_ann_model()
    
    callback_list = [
        tensorflow.keras.callbacks.EarlyStopping( # 성능 향상이 멈추면 훈련을 중지
        monitor='val_loss', # 모델 검증 정확도를 모니터링
        patience = 5
        ),
        tensorflow.keras.callbacks.ModelCheckpoint( # 에포크마다 현재 가중치를 저장
        filepath=op.model_e, #모델 파일 경로 # “model/model_o_h5“ 와 같이 파일명 직접써도 됨
        monitor="val_loss",
        save_best_only=True
        )
    ]
    history = model.fit(X_total,
                        Y_total,
                        epochs=op.num_epochs, 
                        shuffle=True,
                        batch_size=100,
                        validation_split=0.2)
    
    #model.svae(op.model_e)
    return history
```

```python
def showHistoryGraph(history, title) :
    history = history.history
    """Cost/Loss  변화 추이 그래프"""
    plt.plot(history['loss'])
    plt.plot(history['val_loss'])
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Training', 'Verification'], loc=0)
    plt.title(title)
    plt.show()
```

```python
if __name__ == '__main__':
    # 원핫
    history = train_o ()
    showHistoryGraph(history, "one hot encoding" )

    # Embedding
    history = train_e()
    showHistoryGraph(history, "Embedding")
 
```
